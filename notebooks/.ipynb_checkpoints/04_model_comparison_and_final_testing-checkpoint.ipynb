# 04 - Model Comparison and Final Testing

This notebook consolidates final outputs from:
- `02_classification_modeling.ipynb`
- `03_regression_modeling.ipynb`

It generates report-ready summary tables and comparison figures.

```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 200)

PROJECT_ROOT = Path.cwd().parent
TABLE_DIR = PROJECT_ROOT / "reports" / "tables"
FIG_DIR = PROJECT_ROOT / "reports" / "figures"
OUTPUT_DIR = PROJECT_ROOT / "outputs"
MODEL_DIR = PROJECT_ROOT / "models"

COMPARE_TABLE_DIR = TABLE_DIR
COMPARE_FIG_DIR = FIG_DIR / "comparison"
COMPARE_FIG_DIR.mkdir(parents=True, exist_ok=True)
COMPARE_TABLE_DIR.mkdir(parents=True, exist_ok=True)
```

```python
required_files = {
    "classification_metrics": TABLE_DIR / "classification_metrics.csv",
    "regression_metrics": TABLE_DIR / "regression_metrics.csv",
    "classification_baseline": TABLE_DIR / "classification_baseline_metrics.csv",
    "regression_baseline": TABLE_DIR / "regression_baseline_metrics.csv",
    "classification_predictions": OUTPUT_DIR / "predictions" / "classification_predictions.csv",
    "regression_predictions": OUTPUT_DIR / "predictions" / "regression_predictions.csv",
}

missing = [name for name, path in required_files.items() if not path.exists()]
if missing:
    print("Missing files:", missing)
    print("Run notebooks 02 and 03 first, then return here.")
else:
    print("All required files found.")
```

```python
clf_metrics = pd.read_csv(TABLE_DIR / "classification_metrics.csv")
reg_metrics = pd.read_csv(TABLE_DIR / "regression_metrics.csv")

clf_baseline = pd.read_csv(TABLE_DIR / "classification_baseline_metrics.csv")
reg_baseline = pd.read_csv(TABLE_DIR / "regression_baseline_metrics.csv")

clf_pred = pd.read_csv(OUTPUT_DIR / "predictions" / "classification_predictions.csv")
reg_pred = pd.read_csv(OUTPUT_DIR / "predictions" / "regression_predictions.csv")

print("Loaded all result artifacts.")
```

```python
final_summary = pd.DataFrame([
    {
        "task": "classification",
        "primary_metric": "weighted_f1",
        "primary_metric_value": float(clf_metrics.loc[clf_metrics["metric"] == "test_f1_weighted", "value"].iloc[0]),
        "secondary_metric": "accuracy",
        "secondary_metric_value": float(clf_metrics.loc[clf_metrics["metric"] == "test_accuracy", "value"].iloc[0]),
    },
    {
        "task": "regression",
        "primary_metric": "rmse",
        "primary_metric_value": float(reg_metrics.loc[reg_metrics["metric"] == "test_rmse", "value"].iloc[0]),
        "secondary_metric": "r2",
        "secondary_metric_value": float(reg_metrics.loc[reg_metrics["metric"] == "test_r2", "value"].iloc[0]),
    },
])

final_summary
```

```python
final_summary.to_csv(COMPARE_TABLE_DIR / "final_task_summary.csv", index=False)
clf_metrics.to_csv(COMPARE_TABLE_DIR / "final_classification_metrics.csv", index=False)
reg_metrics.to_csv(COMPARE_TABLE_DIR / "final_regression_metrics.csv", index=False)

print("Saved summary tables to:", COMPARE_TABLE_DIR)
```

```python
plt.figure(figsize=(8, 4))
plot_df = clf_baseline.sort_values("val_f1_weighted", ascending=False)
sns.barplot(data=plot_df, x="model", y="val_f1_weighted", palette="viridis")
plt.ylim(0, 1)
plt.title("Classification Baseline Comparison (Validation Weighted F1)")
plt.tight_layout()
plt.savefig(COMPARE_FIG_DIR / "classification_baseline_comparison.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plot_df = reg_baseline.sort_values("rmse", ascending=True)
sns.barplot(data=plot_df, x="model", y="rmse", palette="magma")
plt.title("Regression Baseline Comparison (Validation RMSE)")
plt.tight_layout()
plt.savefig(COMPARE_FIG_DIR / "regression_baseline_comparison.png", dpi=300)
plt.show()
```

```python
clf_match_rate = (clf_pred["actual"] == clf_pred["predicted"]).mean()
reg_error_abs_mean = (reg_pred["actual"] - reg_pred["predicted"]).abs().mean()

extra_checks = pd.DataFrame([
    {"check": "classification_match_rate", "value": float(clf_match_rate)},
    {"check": "regression_mean_absolute_error_from_predictions", "value": float(reg_error_abs_mean)},
])

extra_checks.to_csv(COMPARE_TABLE_DIR / "final_prediction_checks.csv", index=False)
extra_checks
```
