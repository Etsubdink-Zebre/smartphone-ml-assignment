# 05 - Deployment Demo

This notebook shows a simple local deployment-style test using saved models.

Goal:
- Load trained classification and regression models
- Run one sample prediction
- Confirm deployment is working

````python
from pathlib import Path

import joblib
import pandas as pd

PROJECT_ROOT = Path.cwd().parent
CLASSIFICATION_MODEL_PATH = PROJECT_ROOT / "models" / "classification" / "best_classification_model.pkl"
REGRESSION_MODEL_PATH = PROJECT_ROOT / "models" / "regression" / "best_regression_model.pkl"

print("Classification model exists:", CLASSIFICATION_MODEL_PATH.exists())
print("Regression model exists:", REGRESSION_MODEL_PATH.exists())
````

````python
clf_model = joblib.load(CLASSIFICATION_MODEL_PATH)
reg_model = joblib.load(REGRESSION_MODEL_PATH)

print("Models loaded successfully")
````

````python
sample_input = pd.DataFrame([
    {
        "age": 22,
        "gender": "Male",
        "social_media_hours": 4.2,
        "gaming_hours": 1.5,
        "work_study_hours": 5.0,
        "sleep_hours": 6.8,
        "notifications_per_day": 180,
        "app_opens_per_day": 110,
        "weekend_screen_time": 7.5,
        "stress_level": "Medium",
        "academic_work_impact": "Yes",
        "daily_screen_time_hours": 6.2,
        "social_gaming_total_hours": 5.7,
        "notifications_per_open": 180 / 110,
    }
])

sample_input
````

````python
predicted_addiction_level = clf_model.predict(sample_input)[0]
predicted_daily_screen_time = reg_model.predict(sample_input)[0]

print("Predicted addiction level:", predicted_addiction_level)
print("Predicted daily screen time (hours):", round(float(predicted_daily_screen_time), 2))
````
